```python
import pandas as pd
import os
```

```python
## concat all files
df = pd.DataFrame()
filenames = os.listdir("./data")
for file in filenames:
    temp_df = pd.read_csv(f"./data/{file}")
    df = pd.concat([temp_df, df])
df
```

```python
def converter(df, open_column = "Open", close_column = "Close"):
    ## if type is str then convert to float
    if type(df[open_column].iloc[0]) == str and type(df[close_column].iloc[0]) == str:
        try:
            df[open_column] = df[open_column].astype(float)
            df[close_column] = df[close_column].astype(float)
        except ValueError:
            df[open_column] = df[open_column].str.split(".", 1)
            df[close_column] = df[close_column].str.split(".", 1)
            for i in range(0, df.shape[0]):
                df.loc[i, open_column] = df.loc[i, open_column][0].replace(",",".")
                df.loc[i, close_column] = df.loc[i, close_column][0].replace(",",".")
            df[open_column] = df[open_column].astype(float)
```

```python

def gapchecker(df, threshold=3, open_column="Open", close_column="Close"):
# Threshold must be percent like %3
    df["GAP"] = False          
    for i in reversed(range(1, df.shape[0])):
        open_val = df[open_column].iloc[i - 1]
        close_val = df[close_column].iloc[i]
        prc_change = pd.Series([float(close_val), float(open_val)]).pct_change()     
        if abs(prc_change[1]) * 100 > float(threshold):
            df["GAP"].iloc[i - 1] = True
            #print(prc_change[1], close_val, open_val)
            
    
 
```

```python
df.reset_index(inplace=True)
df.drop(["High", "Low", "index"], axis=1, inplace=True)
```

```python
threshold = 2.14
converter(df)
gapchecker(df, threshold)
result = df.groupby(["GAP"]).count()
sum = result["Date"][0] + result["Date"][1]
false = result["Date"][0] / sum
true = result["Date"][1] / sum
print(f"for %{threshold} change threshold, Can be gapped: %{(true * 100).round()}, Can't Gapped %{(false * 100).round()}")
```
